```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 3차 시스템
G = ct.TransferFunction([1], np.polymul([1, 1], np.polymul([1, 2], [1, 5])))

print("Transfer Function G(s):")
print(G)

print("\nPoles:")
print(ct.poles(G))

print("\nZeros:")
print(ct.zeros(G))
```

```python
poles = ct.poles(G)

# 실수부 기준으로 원점에 가장 가까운 pole 찾기
dominant_pole = poles[np.argmax(np.real(poles))]

print("Dominant pole:", dominant_pole)
```

```python
# Step response
t, y = ct.step_response(G)

plt.figure(figsize=(8, 4))
plt.plot(t, y, label='3rd-order step response')
plt.axhline(y[-1], linestyle='--', linewidth=1, label='Final value')
plt.xlabel('Time')
plt.ylabel('Output')
plt.title('3rd-Order System Step Response')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# (s+3)(s^2+2s+5)
den = np.polymul([1, 3], [1, 2, 10])
G2 = ct.TransferFunction([1], den)
```

```python
print("Transfer Function G2(s):")
print(G2)

print("\nPoles:")
print(ct.poles(G2))

print("\nZeros:")
print(ct.zeros(G2))
```

```python
t, y = ct.step_response(G2)

plt.figure(figsize=(8, 4))
plt.plot(t, y, label='3rd-order with complex poles')
plt.axhline(y[-1], linestyle='--', linewidth=1, label='Final value')
plt.xlabel('Time')
plt.ylabel('Output')
plt.title('3rd-Order System with Real Pole + Complex Pole Pair')
plt.grid(True)
plt.legend()
plt.show()
```

```python
poles = ct.poles(G2)
zeros = ct.zeros(G2)

plt.figure(figsize=(6, 5))
plt.axhline(0, linewidth=1)
plt.axvline(0, linewidth=1)

# poles
plt.plot(np.real(poles), np.imag(poles), 'x', markersize=10, label='Poles')

# zeros
if len(zeros) > 0:
    plt.plot(np.real(zeros), np.imag(zeros), 'o', markersize=8, fillstyle='none', label='Zeros')

plt.xlabel('Re(s)')
plt.ylabel('Im(s)')
plt.title('Pole-Zero Map')
plt.grid(True)
plt.legend()
plt.show()
```

```python
den1 = [1, 3]        # s + 3
den2 = [1, 2, 5]     # s^2 + 2s + 5
den3 = [1, 2, 10]    # s^2 + 2s + 10

den = np.polymul(np.polymul(den1, den2), den3)

G5 = ct.TransferFunction([1], den)
```

```python
print("Transfer Function G5(s):")
print(G5)

print("\nPoles:")
print(ct.poles(G5))

print("\nZeros:")
print(ct.zeros(G5))
```

```python
s = ct.TransferFunction.s

G5 = 1 / ((s + 3) * (s**2 + 2*s + 5) * (s**2 + 2*s + 10))

print(G5)
print("Poles:", ct.poles(G5))
print("Zeros:", ct.zeros(G5))
```

```python
t, y = ct.step_response(G5)

plt.figure(figsize=(8, 4))
plt.plot(t, y, label='5th-order with complex poles')
plt.axhline(y[-1], linestyle='--', linewidth=1, label='Final value')
plt.xlabel('Time')
plt.ylabel('Output')
plt.title('5th-Order System with Real Pole + Complex Pole Pair')
plt.grid(True)
plt.legend()
plt.show()
```
